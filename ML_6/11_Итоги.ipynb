# **11. Итоги**

✍ Итак, реальные данные в большинстве случаев неполны, зашумлены и противоречивы.

С экспоненциально растущим объёмом генерации данных и увеличением числа разнородных источников вероятность сбора аномальных или неверных данных достаточно высока.

Но только высококачественные данные могут в итоге дать точные модели и в конечном счёте точные прогнозы. Следовательно, крайне важно обрабатывать данные для достижения максимально возможного качества.

### → Поэтому мы настоятельно рекомендуем вам самостоятельно прогнать все ноутбуки, представленные в модуле, и обратить особое внимание на материал про кодировку, обработку пропусков и аномалий, так как данные разделы являются неотъемлемой частью любого пайплайна машинного обучения.

В этом модуле мы повторили основные этапы предобработки данных и научились применять их на практике.

**ВЫ УЗНАЛИ:**
+ чем отличаются разные методы масштабирования признаков и когда уместно применять конкретные методы;
+ зачем нужна трансформация распределений признаков;
+ какую информацию можно извлечь из широты и долготы;
+ какие алгоритмы машинного обучения более чувствительны к предобработке данных.

**ВЫ НАУЧИЛИСЬ:**
+ осуществлять нужную кодировку категориальных признаков в зависимости от их природы и задачи;
+ использовать географические данные для визуальной локализации объектов на карте.
